# Orbit Wars Demo
This notebook demonstrates how to run and visualize the Orbit Wars environment.

```python
from kaggle_environments import make

# Create environment
env = make("orbit_wars")

# Run with random vs starter agent
state = env.run(["starter", "random"])

# Render in the notebook
env.render(mode="ipython", width=800, height=600)
```

```python
# Run a 4-player game
env.run(["starter", "random", "random", "random"])

# Render in the notebook
env.render(mode="ipython", width=800, height=600)
```
